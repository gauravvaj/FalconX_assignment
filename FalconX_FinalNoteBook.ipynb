### Assignment 2 

```python
# Scatter plot for understanding price trends with respect to time of 3 days data 

sns.scatterplot(x="timestamp", y="trade_price", marker="x", s=0.5, alpha=0.2, data=data_combine)
```

```python
data_combine["trade_price"].mean()
```

It is mentioned in Assignment 2 that we have 3 basis point difference bewtween our ask or offer price and actual value of the price .

```python
#Offer price is here buy price .
data_combine["buy_price"]= data_combine["trade_price"]-0.03*(data_combine["trade_price"])
```

```python
#Ask price is here sell price.
data_combine["sell_price"]=data_combine["trade_price"]+0.03*(data_combine["trade_price"])
```

```python
data_combine.head()
```

```python
#If we sell this is total price 
data_combine["sell_trading_price"]=data_combine.sell_price*data_combine.trade_size
```

```python
# If we buy this is total price.
data_combine["buy_trading_price"]=data_combine.buy_price*data_combine.trade_size
data_combine.head()
```

```python
data_combine.head()
```

It is mentioned that valid trades are between 1000$(i.e.is our offer or bid) and 5000$(hedge value )

```python
data_refine1=data_combine[(data_combine.buy_trading_price>=1000) & (data_combine.buy_trading_price<=5000)]
```

```python
data_refine1.head()
```

```python
data_refine1.shape
```

It could be the case where buy side is less than 5000$ and sell side is above 5000$ so figuring out those rows .

```python
data_refine2=data_refine1[data_refine1["sell_trading_price"]>5000]
```

```python
data_refine2.head()
```

```python
data_refine2.shape
```

```python
data_refine2.tail()
```

Deleting out those rows which were taken in above dataframe where sell_trading_price was greater than 5000.

```python
data_refine1.drop(data_refine1[data_refine1['sell_trading_price'] > 5000].index, inplace = True)
```

```python
data_refine1.shape
```

Inserting a Column "final" which shows the trade  was on buy side or sell side at that instance . This is calulated on the bassis that if the actual trade price was more than average of 3 days trade price then it is postion of selling whereas if the actual trade price is lower than the average of 3 days trade price then it is position of buying. 

Also the "Final" is True for Sell and False for Buy. 
That shows that was sell or buy instance .
True=Sell
False=Buy

```python
data_refine1["final"] = np.where(data_refine1["trade_price"]>data_refine1["trade_price"].mean(), True, False)
```

```python
data_refine1
```

Same applied to dataframe where sell_trading _price was greater than 5000$.

```python
data_refine2["final"] = np.where(data_refine2["trade_price"]>data_refine2["trade_price"].mean(), True, False)
```

```python
data_refine2
```

```python
# true=Sell and false=buy
```

```python
data_refine2.final.value_counts()
```

Taking only those rows into a new dataframe data_refine3 from data_refine2 where it was only buy as if sell is greater than 5000$ we cant consider those rows. 

```python
data_refine3=data_refine2[data_refine2.final==False]
```

```python
data_refine3.shape # having data of only buy 
```

```python
frames=[data_refine1,data_refine3]        # Data of all 3 days combined .
data_refine4=pd.concat(frames) 
```

data_refine4 is the Final dataset with combining data_refine1 and data_refine3 rows .

```python
data_refine4 # Final Dataset.
```

This is the final dataset for Assignment2 according to the rules provided and here are the trades which happened in 3 days . Some of them were on sell side some were on the buy side . How much Profit or Loss we got by trading on each day calculated  in Assignment3 .
